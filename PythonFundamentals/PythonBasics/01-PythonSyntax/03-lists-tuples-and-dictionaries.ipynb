#Lists

```python
my_list = ["a","b","c","d"]

#A value can be accessed by indicating the index.
print(my_list[0])

#We can specify that we only want the first two elements of the list.
print(my_list[:2])

#We can also reverse the list.
print(my_list[::-1])

#Multiplying the amout and replacing the original list
my_list = my_list * 5
print(my_list)

my_list = ["a","b","c","d"]

#Adding  data to the list
my_list.append("Add Text")
print(my_list)
```

We can put any data type in a list

```python
var = "Hello this is a string"
other_list = [200.1, "hi", ["List inside another list","Interesting right!"], 90, var]
print(other_list)
```

We can also merge two lists

```python
x = [1,2,3,4]
y = [4,5,6,7]
x.extend(y)
print(x)

#OR
x = [1,2,3,4]
y = [4,5,6,7]
xy = x + y
print(xy)
```
