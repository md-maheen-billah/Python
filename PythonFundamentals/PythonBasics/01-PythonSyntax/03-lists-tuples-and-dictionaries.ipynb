#Lists

```python
my_list = ["a", "b", "c", "d"]
```

###A value can be accessed by indicating the index.

```python
print(my_list[0])
```

```python
#We can specify that we only want the first two elements of the list.
print(my_list[:2])
```

```python
#We can also reverse the list.
print(my_list[::-1])
```

```python
#Multiplying the amout and replacing the original list
my_list = my_list * 5
print(my_list)
```

```python
my_list = ["a","b","c","d"]

#Adding  data to the list
my_list.append("Add Text")
print(my_list)
```

#We can put any data type in a list

```python
var = "Hello this is a string"
other_list = [200.1, "hi", ["List inside another list","Interesting right!"], 90, var]
print(other_list)
```

#We can also merge two lists

```python
x = [1,2,3,4]
y = [4,5,6,7]
x.extend(y)
print(x)

#OR
x = [1,2,3,4]
y = [4,5,6,7]
xy = x + y
print(xy)
```

#Tuples

```python
tuple_example = ("Bangladesh", "Belgium", "Bhutan")
tuple_example.append("Australia")
print(tuple_example) #there will be an error since we can't use append on tuple
```

#Tuples Vs List

If you need to process a collection that will not change, we will use tuples. 

If we know that a particular collection will have to be changed, then we weill use the lists.

#Dictionaries

Each element of a dictionary consists of a pair of objects: an index and a value, separated by a colon. The indexes are called keys. We use keys to extract a value, therefore the order the elements are present inside a distionary is irrelevent.

```python

heroes = {}
heroes["Batman"] = "Bruce Wayne"
heroes["Superman"] = "Clark Kent"
```

```python
print(heroes) #printing the whole dictionary
```

```python
print(heroes["Batman"]) #printing a specific key
print(heroes["Superman"])
```

```python
heroes["Spiderman"] = "Peter Parker" #we can easily add more keys to the list, no particular method needed
print(heroes)
```
