#Python (high-level language, has a dynamic variable typing)

```python
first_name = "Maheen"  # String
last_name = "Billah"  # String
age = 28  # Integer
weight = 85.5  # Float
becode_student = True  # Boolean
student = [first_name,last_name,age,weight,becode_student] #list
print(student)

"""Note-
True and False are written in uppercase
List is an array, but can store values of different types
"""
```

#Python function type( )

```python
print(first_name,type(first_name))
print(last_name,type(last_name))
print(age,type(age))
print(weight,type(weight))
print(becode_student,type(becode_student))
print(student, type(student))
```
