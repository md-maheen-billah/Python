#Conditions

###if

```python
age = 28
if age >= 18:
    print("You are an adult!")
```

###elif

```python
age = 17
if age >= 18:
    print("You are an adult!")
elif age == 17:
    print("You are one year short from being an")
```

###else

```python
age = 16
if age >= 18:
    print("You are an adult")
else:
    print("You are not an adult")
```

#Logical operators

###Operator and

```python
student1 = "John"
student2 = "Eric"

if student1 == "John" and student2 == "Eric":
    print(f"You are {student1} and {student2}")
else:
    print("Welcome anonymous!")
```

###Operator or

```python
student1 = "John"
student2 = "Eric"

if student1 == "John" or student2 == "Eric":
    print(f"You are either {student1} or {student2}")
else:
    print("Welcome anonymous!")
```

###The "in" operator

```python
name = "Maheen"
team = ["Maheen","Alex"]
if name in team:
    print("You are in the team!")
else:
    print("You are not in the team")
```
