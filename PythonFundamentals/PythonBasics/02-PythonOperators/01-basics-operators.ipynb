#Arithmetic Operators

```python
a = 10
b = 20
```

###Addition

```python
addition = a + b
print(addition)
```

###Subtraction

```python
subtraction = b-a
print(subtraction)
```

###Multiplication

```python
multiplication =a*b
print(multiplication)
```

###Division

Note that in case of division, python returns a flaot

```python
division = b/a
print(division)
```

###Floor Division

To obtain an integer, we will be using floor division

```python
floor_division = b//a
floor_division_bis = 31//10

print("floor_division : ", floor_division)
print("floor_division_bis : ", floor_division_bis) #Outputs 3 instead of 3.1
```

###Modulus

```python
modulus = b%a
print("modulus : ", modulus)
modulus_bis = 21%2
print("modulus_bis : ", modulus_bis)
```

###Exponent

```python
power_of_two = a ** 2
print(power_of_two)
```

#Comparison Operators

```python
a = 10
b = 20
```

###Equals

```python
print(a == b)
print( a == 10)
```

###Not Equals

```python
print(a != b)
print(a != 10)
```

###Bigger than

```python
print(a > b)
print(b > a)
```

###Smaller than

```python

print(a < b)
print(b < a)
     
```

###Smaller or equal than

```python

print(a <= b)
print(b <= a)
```

#Assignment Operators

```python
a = 10
name = "Alan Turing"
print(a)
print(name)
```

###Add AND (+=)

```python
a += 10
name += " is a good mathematician"
print(a)
print(name)
```

###We can only add variables of the same type

```python
name +=a
a += name
```

###Subtract AND (-=)

```python
a = 20
a -= 10
print(a)
```

###This operator does not work with strings, unlike the += operator.

```python
name -= "Alan Turing"
```

###Multiply AND (*=)

```python
a = 10
a *= 10
print(a)

text = "Alan Turing"
text *= 10
print(text)
```

###Divide AND (/=)

This operator does not work with strings, unlike the *= operator.

```python
a = 100
a /= 10
print(a)
```

###Modulus AND (=)

This operator does not work with strings.

```python
a = 100
a %= 3
print(a)
```

###Exponent AND (**=)

```python
a = 2
a **= 3  # ( 2 * 2 * 2)
print(a)
```

###Floor Division AND (//=)

```python
a = 20
a //= 3
print(a)
```
