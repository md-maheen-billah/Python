#Loops and iteration

###While

The while loop is used when it is impossible to determine the exact number of iterations of the loop in advance.

```python
i = 1
while i<=10:
    print(i)
    i += 1
```

###For

A for loop is used for iterating over a sequence

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits: #works with strings
    print(x)
```

```python
for x in "maheen": #works with strings
    print(x)
```

###Range

```python
for i in range(10):
    print(i)
```

```python
for i in range(2**3):
    print(i)
```

###We can specify the starting number.

```python
for i in range(4, 10):
    print(i)
```

###And we can also specify the interval between each.

```python
for i in range(0, 10, 2):
    print(i)
```

#Continue and Break

###Continue

```python
for i in range(4):
    print("Start of iteration", i)
    print("Hello")
    if i < 2:
        continue
    print("End of iteration",i)
print("After the loop")
```

###Break

```python
for i in range(10):
    print("Start of iteration",i)
    print("Hello")
    if i == 2:
        break
    print("End of iteration",i)
print("After the loop")
```

#Else

```python
i = 1
while i < 10:
    print(i)
    i += 1
else:
    print("Finally Finished!, i= ",i)
```

```python
for i in range(1, 10):
    print(i)
else:
    print("Finally finished!, i=",i)
```

###With break, the else clause is not executed.

```python
for i in range(1, 10):
    print(i)
    break
else:
    print("Finally finished!, i =", i)
```
